```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
```

```python
from litellm import completion
```

```python
# OpenAI via LiteLLM
if os.environ.get("OPENAI_API_KEY"):
    result = completion(
        model="openai/gpt-4o-mini",
        messages=[{
            "role": "user",
            "content": "Give me 3 fun facts about Vietnam."
        }],
        max_tokens=150,
    )
    print("LiteLLM (OpenAI) text:", result.choices[0].message.content)
else:
    print("ℹ️ Skipping OpenAI text via LiteLLM (set OPENAI_API_KEY)")
```

```python
# Gemini via LiteLLM
img_url = ""
if os.environ.get("GEMINI_API_KEY") or os.environ.get("GOOGLE_API_KEY"):
    result = completion(
        model="vertex_ai/gemini-1.5-pro",
        messages=[{
            "role":"user",
            "content":[
                {"type": "text", "text": "Describe this image in one sentence."},
                {"type": "image_url", "image_url": {"url": img_url}},
            ]
        }],
        max_tokens=120,
    )
    print("LiteLLM (Gemini) vision:", result.choices[0].message.content)
else:
    print("ℹ️ Skipping Gemini vision via LiteLLM (set GEMINI_API_KEY/GOOGLE_API_KEY)")
```

```python
# Mistral via LiteLLM
if os.environ.get("MISTRAL_API_KEY"):
    result = completion(
        model="mistral/pixtral-12b",
        messages=[{
            "role":"user",
            "content":[
                {"type": "text", "text": "Describe this image in one sentence."},
                {"type": "image_url", "image_url": {"url": img_url}},
            ]
        }],
        max_tokens=120,
    )
    print("LiteLLM (Mistral) vision:", result.choices[0].message.content)
else:
    print("ℹ️ Skipping Pixtral vision via LiteLLM (set MISTRAL_API_KEY)")
```
