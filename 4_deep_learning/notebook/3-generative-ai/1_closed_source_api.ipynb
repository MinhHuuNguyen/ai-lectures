```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
```

## OpenAI


```python
from openai import OpenAI
```

```python
openai_key = os.environ.get("OPENAI_API_KEY")
```

```python
o_client = OpenAI(api_key=openai_key)
```

```python
result = o_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{
        "role": "user",
        "content": "Give me 3 fun facts about Vietnam."
    }],
    max_tokens=200,
)
result
```

```python
result.choices[0].message.content
```

```python
img_url = ""
result = o_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{
        "role":"user",
        "content":[
            {"type": "text", "text": "Describe this image in one sentence."},
            {"type": "image_url", "image_url": {"url": img_url}},
        ]
    }],
    max_tokens=150,
)
result
```

```python
result.choices[0].message.content
```

## Anthropic


```python
from anthropic import Anthropic
```

```python
anthropic_key = os.environ.get("ANTHROPIC_API_KEY")
```

```python
a_client = Anthropic(api_key=anthropic_key)
```

```python
result = a_client.messages.create(
    model="claude-3-haiku-20240307",
    messages=[{
        "role": "user",
        "content": "Give me 3 fun facts about Vietnam."
    }],
    max_tokens=200,
)
result
```

```python
result.content[0].text
```

```python
img_url = ""
result = a_client.messages.create(
    model="claude-3-haiku-20240307",
    max_tokens=150,
    messages=[{
        "role":"user",
        "content":[
            {"type": "text", "text": "Describe this image in one sentence."},
            {"type":"image", "source":{"type": "url", "url": img_url}},
        ],
    }],
)
result
```

```python
result.content[0].text
```

## Google Gemini

```python
from google import genai as google_genai
from google.genai import types as genai_types
```

```python
gemini_key = os.environ.get("GEMINI_API_KEY") or os.environ.get("GOOGLE_API_KEY")
```

```python
g_client = google_genai.Client(api_key=gemini_key)
```

```python
result = g_client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="Give me 3 fun facts about Vietnam.",
)
result
```

```python
result.text
```

```python
img_url = ""
result = g_client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=[
        genai_types.Part.from_text("Describe this image in one sentence."),
        genai_types.Part.from_uri(file_uri=img_url, mime_type="image/jpeg"),
    ],
)
result
```

```python
result
```
