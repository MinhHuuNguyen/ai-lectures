```python
import os
```

```python
from huggingface_hub import InferenceClient
```

```python
hf_token = os.environ.get("HUGGINGFACEHUB_API_TOKEN")
```

```python
client = InferenceClient(model="google/gemma-2-2b-it", token=hf_token)
```

```python
text = client.text_generation("Give me 3 fun facts about Vietnam.", max_new_tokens=120)
text
```

```python
client = InferenceClient(model="nlpconnect/vit-gpt2-image-captioning", token=hf_token)
```

```python
img_url = ""
caption = client.image_to_text(image=img_url)
caption
```
