```python
import torch
```

```python
# Model
model = torch.hub.load("ultralytics/yolov5", "yolov5s")  # or yolov5n - yolov5x6, custom
model
```

```python
# Images
img = "yolov5/data/images/bus.jpg"  # or file, Path, PIL, OpenCV, numpy, list
```

```python
# Inference
results = model(img)
results
```

```python
# Results
results.print()  # or .show(), .save(), .crop(), .pandas(), etc.
```
