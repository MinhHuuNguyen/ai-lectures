# POST

```python
request_dict = {
    'image_path': 'https://raw.githubusercontent.com/ultralytics/yolov5/master/data/images/bus.jpg'
}
```

```python
response = requests.post('http://localhost:1504/predict_path', json=request_dict)
```

```python
response
```

```python
response.json()
```

```python
image = Image.open('yolov5/data/images/bus.jpg')
image
```

```python
image_bytes = BytesIO()
image.save(image_bytes, format="JPEG")
```

```python
image_bytes
```

```python
response = requests.post('http://localhost:1504/predict_file', data=image_bytes.getvalue())
```
