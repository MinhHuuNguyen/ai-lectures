```python
import numpy as np
```

```python
# 1. Viết chương trình NumPy để chuyển đổi một danh sách các giá trị số thành mảng NumPy một chiều.
```

```python
lst = [1, 2, 3, 4, 5]
arr = np.array(lst)
```

```python
print("Mảng NumPy 1 chiều:", arr)
```

```python
# 2. Viết chương trình NumPy để tạo ma trận 3x3 với các giá trị từ 2 đến 10.
```

```python
matrix = np.arange(2, 11).reshape(3, 3)
```

```python
print("Ma trận 3x3 từ 2 đến 10:\n", matrix)
```

```python
# 3. Viết chương trình NumPy để đảo ngược một mảng (phần tử đầu tiên trở thành cuối cùng).
```

```python
arr = np.array([10, 20, 30, 40, 50])
```

```python
print("Mảng ban đầu:", arr)
```

```python
reversed_arr = arr[::-1]
```

```python
print("Mảng sau khi đảo ngược:", reversed_arr)
```

```python
# 4. Viết chương trình NumPy để chuyển đổi mảng sang kiểu `float`.
```

```python
arr = np.array([1, 2, 3, 4])
```

```python
print("Mảng ban đầu:", arr, arr.dtype)
```

```python
float_arr = arr.astype(float)
```

```python
print("Mảng float:", float_arr, float_arr.dtype)
```

```python
float_arr = arr.astype(np.float16)
```

```python
print("Mảng float:", float_arr, float_arr.dtype)
```

```python
# 5. Viết chương trình NumPy để tạo mảng 2D có giá trị `1` ở viền và `0` ở bên trong.
```

```python
arr = np.ones((5, 5), dtype=int)
```

```python
print("Mảng 2D ban đầu:\n", arr)
```

```python
arr[1:-1, 1:-1] = 0
```

```python
print("Mảng 2D với viền là 1 và bên trong là 0:\n", arr)
```

```python
# 6. Viết chương trình NumPy để tạo ma trận 8x8 và điền theo mẫu bàn cờ (checkerboard).
```

```python
arr = np.zeros((8, 8), dtype=int)
```

```python
print("Mảng 2D ban đầu:\n", arr)
```

```python
arr[1::2, ::2] = 1
arr[::2, 1::2] = 1
```

```python
print("Mẫu bàn cờ 8x8:\n", arr)
```

```python
# 7. Viết chương trình NumPy để chuyển đổi giá trị nhiệt độ từ độ C sang độ F. (Công thức: `F = C * 9 / 5 + 32`)
```

```python
celsius = np.array([0, 20, 37, 100])
```

```python
print("Độ C:", celsius)
```

```python
fahrenheit = celsius * 9 / 5 + 32
```

```python
print("Độ F:", fahrenheit)
```

```python
# 8. Viết chương trình NumPy để lấy chỉ số (index) của các phần tử đã được sắp xếp trong mảng.
```

```python
arr = np.array([40, 10, 30, 20])
```

```python
print("Mảng ban đầu:", arr)
```

```python
sorted_indices = np.argsort(arr)
```

```python
print("Chỉ số sau khi sắp xếp:", sorted_indices)
```
