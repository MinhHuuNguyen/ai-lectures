### Câu 1:

- Có bao nhiêu sản phẩm:
    - trong danh sách 1?
    - trong danh sách 2?
    - trong chỉ danh sách 1 (Liệt kê tên các sản phẩm đó)?
    - trong chỉ danh sách 2 (Liệt kê tên các sản phẩm đó)?
    - trong cả hai danh sách (Liệt kê tên các sản phẩm đó)?
- Tạo tệp csv mới chỉ chứa các sản phẩm duy nhất từ hai danh sách (Đặt tên tệp là items.csv).

```python
print(f'There are {len(item_list_1_df)} items in list 1')
print(f'There are {len(item_list_2_df)} items in list 2')
```

```python
merged_df = pd.merge(
    item_list_1_df,
    item_list_2_df,
    on='item_id',
    how='outer',
    indicator=True
)
merged_df
```

```python
def process_df(_df):
    _df = _df.dropna(axis=1).drop(columns='_merge')
    
    rename_dict = dict()
    for col_name in _df.columns:
        if '_x' in col_name or '_y' in col_name:
            rename_dict[col_name] = col_name[:-2]
    
    new_df = _df.rename(columns=rename_dict)
    return new_df
```

```python
item_only_list_1_df = merged_df.loc[merged_df._merge == 'left_only']
item_only_list_1_df
```

```python
item_only_list_1_df = process_df(item_only_list_1_df)
item_only_list_1_df
```

```python
item_only_list_2_df = merged_df.loc[merged_df._merge == 'right_only']
item_only_list_2_df
```

```python
item_only_list_2_df = process_df(item_only_list_2_df)
item_only_list_2_df
```

```python
item_both_list_df = merged_df.loc[merged_df._merge == 'both']
item_both_list_df
```

```python
item_both_list_df = process_df(item_both_list_df)
item_both_list_df
```

```python
item_both_list_df = item_both_list_df.loc[:,~item_both_list_df.columns.duplicated()]
item_both_list_df
```

```python
print(f'There are {len(item_only_list_1_df)} items in only list 1')
print(f'There are {len(item_only_list_2_df)} items in only list 2')
print(f'There are {len(item_both_list_df)} items in both list 1 and list 2')
```

```python
items_df = pd.concat(
    [item_only_list_1_df, item_only_list_2_df, item_both_list_df]
)
items_df.to_csv('items.csv', index=False)
items_df
```

```python
def check_digits(name):
    for char in name:
        if char.isdigit():
            return True
    return False
```

### Câu 4:

- Sản phẩm nào có giá cao nhất trong mỗi năm?
- Sản phẩm nào có giá thấp nhất trong mỗi năm?
- Tính giá trung bình của từng sản phẩm trong mỗi năm.

```python
items_df
```

```python
sales_item_name_df = pd.merge(
    sales_df, items_df, on='item_id', how='left'
)
sales_item_name_df
```

```python
sales_item_name_df['year'] = sales_item_name_df.date.apply(
    lambda x: x.split('.')[-1]
)
sales_item_name_df
```

```python
def find_highest_lowest_price(df, year):
    sales_year_df = df.loc[df.year == year]

    max_price = np.max(sales_year_df.item_price)
    highest_df = sales_year_df.loc[sales_year_df.item_price == max_price]

#     sales_year_df = sales_year_df.loc[sales_year_df.price > 0]
    min_price = np.min(sales_year_df.item_price)
    lowest_df = sales_year_df.loc[sales_year_df.item_price == min_price]

    average_price = np.average(sales_year_df.item_price)

    return highest_df, lowest_df, average_price
```

```python
year_list = sales_item_name_df.year.unique()
year_list
```

```python
'''
price_dict = {
    '2013': {
        'highest': highest_df,
        'lowest': lowest_df,
        'average_price': average_price
    },
    '2014': {
        'highest': highest_df,
        'lowest': lowest_df,
        'average_price': average_price
    },
    '2015': {
        'highest': highest_df,
        'lowest': lowest_df,
        'average_price': average_price
    }
}
'''
```

```python
price_dict = dict()

for year in year_list:
    highest_df, lowest_df, average_price = find_highest_lowest_price(
        sales_item_name_df, year)

    price_dict[year] = dict()
    price_dict[year]['highest'] = highest_df
    price_dict[year]['lowest'] = lowest_df
    price_dict[year]['average_price'] = average_price
```
