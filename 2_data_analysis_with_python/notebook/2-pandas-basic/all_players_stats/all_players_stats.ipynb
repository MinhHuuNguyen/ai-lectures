```python
# Câu 1: Vẽ biểu đồ cột thể hiện số bàn thắng của các cầu thủ CLB Manchester United ghi được.
```

```python
mu_df = df.loc[df['Team'] == 'Manchester United']
mu_df
```

```python
plt.figure(figsize=(20, 6))
plt.bar(mu_df['Player'], mu_df['Goals'], color='r')
plt.xticks(rotation=90)
plt.xlabel('Player name')
plt.ylabel('Number of goals')
plt.title('GOALS')
plt.show()
```

```python
mu_sorted_df = mu_df.sort_values(by='Goals', ascending=False)
mu_sorted_df
```

```python
plt.figure(figsize=(20, 6))
plt.bar(mu_sorted_df['Player'], mu_sorted_df['Goals'], color='r')
plt.xticks(rotation=90)
plt.xlabel('Player name')
plt.ylabel('Number of goals')
plt.title('GOALS')
plt.show()
```

```python
# Câu 2: Vẽ biểu đồ tròn thể hiện tỷ lệ số bàn thắng của CLB Manchester United theo từng vị trí thi đấu.
# Với các cầu thủ không ghi bàn thì không tính vào biểu đồ tròn này.
# Với các cầu thủ thi đấu nhiều vị trí thì lấy vị trí chính (vị trí đứng đầu trong chuỗi Position)
```

```python
mu_goal_df = mu_df.loc[mu_df['Goals'] > 0]
mu_goal_df['MainPosition'] = mu_goal_df.Position.apply(lambda x: x.split('/')[0])
mu_goal_df
```

```python
mu_grouped_goal_df = mu_goal_df.groupby('MainPosition').agg({'Goals': 'sum'}).reset_index()
mu_grouped_goal_df
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    mu_grouped_goal_df['Goals'],
    labels=mu_grouped_goal_df['MainPosition'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Goals by Main Position')
plt.show()
```

```python
# Câu 3: Vẽ biểu đồ cột nằm ngang thể hiện tổng số thẻ vàng của từng CLB.
```

```python
grouped_yellow_df = df.groupby('Team').agg({'YellowCards': 'sum'}).reset_index()
grouped_yellow_df
```

```python
plt.figure(figsize=(10, 6))
plt.barh(grouped_yellow_df['Team'], grouped_yellow_df['YellowCards'], color='orange')
plt.xlabel('Total Yellow Cards')
plt.ylabel('Team')
plt.title('Total Yellow Cards by Team')
plt.show()
```

```python
# Câu 4: Vẽ biểu đồ miền thể hiện tỷ lệ giữa top 3 cầu thủ ghi nhiều bàn thắng nhất của mỗi đội.
# Trong đó: Trục x => tên của các CLB, Trục y => tổng số bàn thắng của top 3 cầu thủ của CLB đó.
```

```python
top_3_goalscorers_df = df.groupby('Team')['Goals'].nlargest(3)
top_3_goalscorers_df
```

```python
type(top_3_goalscorers_df)
```

```python
goals_data = df.groupby('Team')['Goals']

top_1_list, top_2_list, top_3_list, team_name_list = [], [], [], []
for team_name, g_data in goals_data:
    tmp_g_data = g_data.nlargest(3).reset_index(drop=True)
    top_1_list.append(tmp_g_data.iloc[0])
    top_2_list.append(tmp_g_data.iloc[1])
    top_3_list.append(tmp_g_data.iloc[2])
    team_name_list.append(team_name)
```

```python
plt.figure(figsize=(20, 6))
plt.stackplot(team_name_list, top_1_list, top_2_list, top_3_list, colors=['r', 'g', 'b'])
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Number of goals')
plt.title('GOALS')
plt.show()
```

```python
# Câu 5: Vẽ biểu đồ scatter thể hiện mối quan hệ giữa số bàn thắng và số trận thi đấu của các cầu thủ tiền đạo.
```

```python
forward_df = df[df['Position'].str.contains('Forward')]
forward_df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(forward_df['Apearances'], forward_df['Goals'], color='purple')
plt.xlabel('Number of Appearances')
plt.ylabel('Number of Goals')
plt.title('Goals vs Appearances for Forwards')
plt.show()
```

```python
# Câu 6: Vẽ biểu đồ histogram thể hiện phân bố về số lượng bàn thắng của các cầu thủ trong bộ dữ liệu.
# Vẽ biểu đồ box plot và violin plot thể hiện phân bố về số lượng bàn thắng của các cầu thủ trong bộ dữ liệu nhóm theo từng CLB.
```
