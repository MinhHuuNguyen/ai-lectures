```python
import pandas as pd
```

```python
# Hãy thực hiện các thao tác làm sạch dữ liệu dưới đây:  
# 1. Loại bỏ các giá trị duplicates
# 2. Split dữ liệu trong các cột nếu cột đó chứa nhiều giá trị trong 1 ô
# 3. Xóa khoảng trống ở đầu, cuối hoặc không cần thiết giữa các từ
# 4. Đổi kiểu dữ liệu
# 5. Xử lý các dữ liệu bị thiếu bằng cách drop, thay thế bằng mode, mean, … 
```

```python
df = pd.read_csv('brasil-real-estate-1.csv')
df
```

```python
# Loại bỏ các giá trị duplicates
# - Giống nhau hoàn toàn: Giống nhau tất cả giá trị trên các cột
# - Giống nhau một phần: Giống nhau một số giá trị trên các cột
```

```python
df = df.drop_duplicates()
df
```

```python
df.drop_duplicates(subset=['region'])
```

```python
# Split dữ liệu trong các cột nếu cột đó chứa nhiều giá trị trong 1 ô
```

```python
def split_(name):
    return name.split('|')[1:-1]
```

```python
df.loc[:, 'split_place_with_parent_names'] = df['place_with_parent_names'].apply(split_)
df
```

```python
df.loc[:, 'number_of_names'] = df['split_place_with_parent_names'].apply(len)
df
```

```python
df['number_of_names'].value_counts()
```

```python
df[['place_1', 'place_2', 'place_3', 'place_4']] = pd.DataFrame(df['split_place_with_parent_names'].tolist(), index=df.index)
df
```

```python
# Xóa khoảng trống ở đầu, cuối hoặc không cần thiết giữa các từ
```

```python
df['property_type'] = df['property_type'].str.strip()
# df['property_type'] = df['property_type'].apply(lambda x: x.strip())
df
```

```python
# Đổi kiểu dữ liệu
```

```python
df.info()
```

```python
df['area_m2'] = df['area_m2'].astype(float)
df
```

```python
df['area_m2'] = df['area_m2'].astype(int)
df
```

```python
df.info()
```

```python
# Xử lý các dữ liệu bị thiếu bằng cách drop, thay thế bằng mode, mean, …
```

```python
# Xử lý cột lat-lon
# Xử lý khuyết thiếu:
# - Cách 1: Drop
# - Cách 2: Điền thêm vào:
#     + Cách 2.1: Điền thêm giá trị đặc biệt (-1, -99, 999999 ...)
#     + Cách 2.2: Điền thêm giá trị thống kê (mean, median, mode ...)
```

```python
df
```

```python
drop_df = df.dropna(subset='lat-lon')
drop_df
```

```python
df['lat-lon'] = df['lat-lon'].fillna('-999999,-999999')
df
```

```python
df[['lat', 'lon']] = df['lat-lon'].str.split(',', expand=True)
df
```

```python
lat_value = df['lat'].astype(float)
lat_mean = lat_value[lat_value != -999999].mean()
lat_mean
```

```python
lon_value = df['lon'].astype(float)
lon_mean = lon_value[lon_value != -999999].mean()
lon_mean
```

```python
df['lat'] = df['lat'].replace("-999999", lat_mean)
df['lon'] = df['lon'].replace("-999999", lon_mean)
df
```
