```python
# pip install --upgrade google-cloud-translate
from google.cloud import translate_v2 as translate

import os
```

# 1. Set up client
## 1.1. Set key environment

```python
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = "gcp_env/vision-csvc-minhhuunguyen.json"
```

## 1.2. Create client

```python
client = translate.Client()
client
```

# 2. Prepare text

```python
text = 'Good Morning. GoodBye. And Hello'
```

# 3. Run Translate

```python
result = client.translate(text, target_language='ja')
result
```

```python
# おはよう。 さようなら。 そしてこんにちは
```

# 4. List Supported Languages

```python
result = client.get_languages()
result
```

# 5. Language classification

```python
result = client.detect_language(text)
result
```

```python
result = client.detect_language('おはよう。 さようなら。 そしてこんにちは')
result
```
