```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
```

```python
IMAGE_FOLDER_PATH = 'brain_tumor_dataset/train/'
```

```python
df = pd.read_csv('brain_tumor_dataset/brain_multi.csv')
df
```

```python
tumor_df = df[df.label == 'tumor']
tumor_df
```

```python
def show_bbox(tumor_df, index=-1, image_name=''):
    if image_name == '' and index == -1:
        raise

    if image_name == '':
        image_path = os.path.join(IMAGE_FOLDER_PATH, tumor_df.iloc[index].image_name)
        bbox = tumor_df.iloc[index][['x1', 'y1', 'x2', 'y2']]

    if index == -1:
        image_path = os.path.join(IMAGE_FOLDER_PATH, image_name)
        bbox = tumor_df.loc[tumor_df.image_name == image_name, ['x1', 'y1', 'x2', 'y2']].squeeze()

    image = cv2.imread(image_path)
    image = cv2.rectangle(
        img=image,
        pt1=(bbox['x1'], bbox['y1']),
        pt2=(bbox['x2'], bbox['y2']),
        color=(255, 0, 0),
        thickness=2
    )
    plt.imshow(image)
    plt.show()
```

```python
show_bbox(tumor_df, index=23)
```
