# 2. Chuẩn bị label

```python
train_df[['image_name', 'label', 'xmin', 'ymin', 'xmax', 'ymax']]
```

```python
def prepare_label_txt(image_name, label, x1, y1, x2, y2, dataset_name):
    img = Image.open(f'brain_tumor_mri_dataset/train/{image_name}')
    img_w, img_h = img.size[0], img.size[1]
    # Convert from Absolute bounding box to Relative bounding box
    x = (x1 + x2) / 2 / img_w
    y = (y1 + y2) / 2 / img_h
    w = (x2 - x1) / img_w
    h = (y2 - y1) / img_h
    image_name = image_name.replace('.jpg', '.txt')
    with open(f'brain_tumor_mri_dataset_coco/labels/{dataset_name}/{image_name}', 'w') as file:
        if label == 'tumor':
            file.write(f'0 {x} {y} {w} {h}')
        else:
            pass
```

```python
for _, row in tqdm(train_df.iterrows()):
    prepare_label_txt(
        row.image_name, row.label, row.xmin, row.ymin, row.xmax, row.ymax, 'train'
    )
```

```python
for _, row in tqdm(val_df.iterrows()):
    prepare_label_txt(
        row.image_name, row.label, row.xmin, row.ymin, row.xmax, row.ymax, 'val'
    )
```

```python
sample_name = os.listdir('./brain_tumor_mri_dataset_coco/labels/val/')[0]
with open(f'./brain_tumor_mri_dataset_coco/labels/val/{sample_name}', 'r') as file:
    lines = file.readlines()
lines
```

# 3. Chuẩn bị file txt đường dẫn

```python
def create_txt_path_file(root_path, dataset_name):
    file_path_list = glob(os.path.join(root_path, dataset_name, '*'))
    file_path_list = [path.replace('brain_tumor_mri_dataset_coco', '.') + '\n' for path in file_path_list]
    with open(f'brain_tumor_mri_dataset_coco/{dataset_name}.txt', 'w') as file:
        file.writelines(file_path_list)
```

```python
create_txt_path_file('brain_tumor_mri_dataset_coco/images/', 'val')
```

```python
create_txt_path_file('brain_tumor_mri_dataset_coco/images/', 'train')
```
